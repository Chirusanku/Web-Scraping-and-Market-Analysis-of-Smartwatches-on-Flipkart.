```python
Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df["Price"] < lower) | (df["Price"] > upper)]
```

###  Price vs Reviews  

Examine the relationship between smartwatch pricing and customer engagement levels to determine whether higher-priced products generate more reviews.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(data=df, x='Price', y='Reviews')

plt.title("Price vs Reviews", fontsize=14, weight='bold')
plt.xlabel("Price")
plt.ylabel("Reviews")
plt.show()
```

```python
top10_brands = df['Brand'].value_counts().head(10).index
```

```python
g = sns.pairplot(
    df,
    vars=['Price_category', 'Ratings', 'Reviews'],
    hue='Price_category',
    diag_kind='hist'
)

g.fig.suptitle("Multivariate Analysis: Price Category vs Ratings vs Reviews",
               y=1.02, fontsize=14)

plt.show()
```
